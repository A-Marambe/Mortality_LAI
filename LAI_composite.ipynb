##### Crating composite LAI images

```python
# import modules
import pandas as pd
import geopandas as gpd
import os

# import modules
import geopandas as gpd
import rasterio as ras
import numpy as np
from datetime import datetime
import glob
from shapely.geometry import box
from shapely.geometry import Polygon
from osgeo import gdal, ogr, osr
from matplotlib import pyplot as plt
```

```python
# list image folders
fold_list = os.listdir('./output/images')
fold_list = [item for item in fold_list if item != '.DS_Store']
fold_list = fold_list
# overide for forelder 57
fold_list = [item for item in fold_list if item=='57']
print(fold_list)
```

```python
def summer_median(image_list):
    '''
    This function coverts list of images to summer median images
    Args:
        param1: image list
    returns:
        type: median value array
    '''
    stacked = np.array(image_list)
    median_image = np.median(stacked, axis=0)#.astype(np.uint8)
    return median_image    
```

```python
# loop through image folders
for item in fold_list:
    print(f'the ploygon folder is {item} *************************')
    # list image folders
    file_list = glob.glob('./output/images/'+item+'/*tif')
    # create a folder to hold composite images for this polygon
    try:
        os.mkdir('./output/composite/'+str(item))
    except:pass
    # selecting images by year within each image folders
    for year in range(2000, 2024):
        print(f'the year is {year}')
        # list to collect 'this' year images
        year_images = []
        for image in file_list:
            # take images only for current year
            image_year = image[-14:-10]
            if image_year != str(year):continue
            else:
                #print(image)
                ds = ras.open(image)
                # read the lai band
                lai = ds.read(1)
                # filter out images with zeros (cloud masked value)
                if np.any(lai == 0, axis=1).all():
                    continue
                else:
                    year_images.append(lai)
                #print(lai)

        print(f'year {year} image object list prepared')
        # create median image
        summer_image = summer_median(year_images)
        # now save them as tiff images
        # composite median image path
        image_path  = './output/composite/'+str(item)+'/'+str(year)+'.tif'
        #save images
        with ras.Env():
            # image profile
            profile = ds.profile
            # update profile
            profile.update(
                dtype= ras.uint8,
                count =1,
                compress = 'lzw'
            )
            with ras.open(image_path, 'w', **profile) as dst:
                dst.write(summer_image.astype(ras.uint8), 1)
                dst.close()
        print(f'Processing completed for {year}')
        print('*********************************')
```
